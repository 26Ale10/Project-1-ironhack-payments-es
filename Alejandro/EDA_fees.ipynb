```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import datetime as dt
```

```python
fees_original = pd.read_csv("..\project_dataset\\fees_data.csv")
fees = fees_original.copy()
```

```python
fees['created_at'] = pd.to_datetime(fees['created_at'])
fees['updated_at'] = pd.to_datetime(fees['updated_at'])
fees['paid_at'] = pd.to_datetime(fees['paid_at'])
fees['from_date'] = pd.to_datetime(fees['from_date'],unit='ns')
fees['to_date'] = pd.to_datetime(fees['to_date'],unit='ns')
```

```python
#Status Accepted.

fees_modidy = fees[['id','cash_request_id','type','status','total_amount','paid_at']]
fees_accepted = fees_modidy[fees_modidy['status']== 'accepted']
fees_accepted.info()
```

```python
# def get_month(x): 
#     return dt.datetime(x.year, x.month,1)

fees_accepted['paid_at_fees_by_YM'] = fees_accepted['paid_at'].dt.to_period('M')
grouping1 = fees_accepted.groupby('cash_request_id')['paid_at_fees_by_YM']
fees_accepted['cohorte_month_fees'] = fees_accepted.groupby('cash_request_id')['paid_at_fees_by_YM'].transform('min')
print(fees_accepted.info())
```
